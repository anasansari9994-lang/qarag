```python
from llama_index.core import StorageContext, load_index_from_storage
from llama_index.core.retrievers import VectorIndexRetriever

store_cotext = StorageContext.from_defaults(persist_dir="./storage_msft")
index = load_index_from_storage(store_cotext)
```

```python
retriever = index.as_retriever(similarity_top_k = 5)
```

```python
%pip install langchain_google_genai
```

```python
def get_context(input_dict):
    question = input_dict["question"]
    nodes = retriever.retrieve(question)
    return "\n".join([node.get_content() for node in nodes])
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough , RunnableLambda
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser

context_retriever = RunnableLambda(get_context)

llm = ChatGoogleGenerativeAI(model="models/gemini-2.5-flash",
google_api_key = api_key)
system_prompt = """
You are a Senior Financial Analyst.

Instructions:
- Answer ONLY using the provided context.
- Do NOT use prior knowledge.
- Do NOT guess or hallucinate.
- If the answer is not explicitly found, say: "Not found in the provided context."

Formatting Rules:
- ALWAYS express large financial numbers in BILLIONS (not millions)
- Format as: "Total Revenue (FY2024): $XXX.X billion"
- Use bullet points for all financial data
- Be precise and concise
- Include numbers, percentages, and dates when available

Context:
{context}

Question:
{question}

Answer:
"""
prompt = ChatPromptTemplate.from_template(system_prompt)

chain = (
    {'context' : context_retriever , 'question' : lambda x: x["question"]}
    |prompt
    |llm
    |StrOutputParser()
)
```

```python
response = chain.invoke({"question": "What is the 2024 revenue for Microsoft?"})
print(response)
```
